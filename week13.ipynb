```python
#실습1: 생일 기온 시각화

import csv
import matplotlib.pyplot as plt

f = open('ta_20210601002300.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)

h, l = [], []
for row in data:
    if row[-2] and row[-1]:
        date = row[0].split('-')
        if date[1] == '04' and date[2] == '09':
            l.append(float(row[-2]))
            h.append(float(row[-1]))        
f.close()

plt.plot(l)
plt.plot(h)
plt.show()
```

```python
#실습2: 최고 기온 시각화

import csv
import matplotlib.pyplot as plt

f = open('ta_20210601002300.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)

jan, aug = [], []
for row in data:
    if row[-1]:
        date = row[0].split('-')
        if date[1] == '01':
            jan.append(float(row[-1]))
        elif date[1] == '08':
            aug.append(float(row[-1]))            
f.close()

plt.hist(aug, color='r', bins=100, label='Aug')
plt.hist(jan, color='b', bins=100, label='Jan')
plt.legend()
plt.show
```
