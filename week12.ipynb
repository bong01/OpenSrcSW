```python
#실습 1

import csv

GAP, MIN, MAX = -999, 999, -999
gap_date, min_date, max_date='', '', ''

f = open('ta_20210524234046.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)

for row in data: 
    if row[-1] !='' and row[-2]!='':
        date = row[0]
        high = float(row[-1])
        low = float(row[-2])
        if high-low > GAP:
            GAP = high-low
            gap_date = date
        if low < MIN:
            MIN = low
            min_date = date
        if high > MAX:
            MAX = high
            max_date = date  
        
f.close()
print("일교차가 가장 큰 날\n"+gap_date+"\n최저기온이 가장 낮은 날\n"+min_date+"\n최고기온이 가장 높은 날\n"+max_date)
```

```python
#실습 2

import csv

f = open('2020년 12월  교통카드 통계자료.csv', 'r', encoding='utf-8')
data = csv.reader(f)
header = next(data)
line={}

for row in data:
    station={}
    station[row[3]] = int(row[4].replace(',',''))+int(row[5].replace(',',''))
    if row[1] in line:
        line[row[1]].update(station)
    else:
        line[row[1]] = station
        
for station in line:
    print('['+station+']'+'\n이용객이 가장 많은 역 순위')
    sorted_list = sorted(line[station].items(), key=lambda x:x[1], reverse=True)
    for i in range(3):
        print(str(i+1)+'. '+sorted_list[i][0]+'('+str(sorted_list[i][1])+'명)')
    print('이용객이 가장 적은 역 순위')
    for j in range(3):
        print(str(j+1)+'. '+sorted_list[-(j+1)][0]+'('+str(sorted_list[-(j+1)][1])+'명)')
    print()    
```
